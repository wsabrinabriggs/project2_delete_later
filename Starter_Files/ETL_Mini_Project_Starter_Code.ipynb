```python
new_columns_list = []
new_columns_list.append(crowdfunding_cols[14].split("&"))
print(new_columns_list)
print(new_columns_list[0])
print(new_columns_list[0][0]) #category
print(new_columns_list[0][1]) #subcategory
```

```python
crowdfunding_values = crowdfunding_info_df.values.tolist()
#print(crowdfunding_values)
print(crowdfunding_values)

```

```python
category_and_subcategory_values_list = []

for item in crowdfunding_values:
    category_and_subcategory_values_list.append(item[14].split("/"))
print(category_and_subcategory_values_list)

```

```python
category_list = []
for item in category_and_subcategory_values_list:
    category_list.append(item[0])
print(category_list)
```

```python
subcategory_list = []
for item in category_and_subcategory_values_list:
    subcategory_list.append(item[1])
print(subcategory_list)
```

```python
# Assign the category and subcategory values to category and subcategory columns.
crowdfunding_info_df2 = crowdfunding_info_df.assign(category=category_list, subcategory=subcategory_list)
crowdfunding_info_df2.head()
```

```python
# Get the unique categories and subcategories in separate lists.
categories = crowdfunding_info_df2['category'].unique().tolist() 
subcategories = crowdfunding_info_df2['subcategory'].unique().tolist()
print(categories)
print(subcategories)
```

```python
# Get the number of distinct values in the categories and subcategories lists.
print(len(categories))
print(len(subcategories))
```

```python
# Create numpy arrays from 1-9 for the categories and 1-24 for the subcategories.
category_ids = np.arange(1, 10)
subcategory_ids = np.arange(1, 25)

print(category_ids)
print(subcategory_ids)
```

```python
# Use a list comprehension to add "cat" to each category_id. 

cat_ids = []
for category_id in category_ids:
    if "cat" not in category_ids:
        cat_ids.append((f"cat{category_id}"))
print(cat_ids)

# cat_ids = [append((f"cat{category_id}") x for x in category_ids if "cat" not in category_ids)]
# cat_ids


# Use a list comprehension to add "subcat" to each subcategory_id.    
scat_ids = []
for subcategory_id in subcategory_ids:
    if "subcat" not in subcategory_ids:
        scat_ids.append((f"subcat{subcategory_id}"))
print(scat_ids)
    
#print(cat_ids)
#print(scat_ids)
```

```python
# Use a list comprehension to add "cat" to each category_id. 

# Use a list comprehension to add "subcat" to each subcategory_id.    

    
print(cat_ids)
print(scat_ids)
```

```python
# Create a category DataFrame with the category_id array as the category_id and categories list as the category name.
category_df = pd.DataFrame(list(zip(cat_ids,categories)), columns=["category_id", "category"])

# Create a category DataFrame with the subcategory_id array as the subcategory_id and subcategories list as the subcategory name. 
subcategory_df = pd.DataFrame(list(zip(scat_ids, subcategories)), columns =["subcategory_id", "subcategory"])
```
